```python
from ucimlrepo import fetch_ucirepo 
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
from random import choices, sample
```

```python
iris = fetch_ucirepo(id=53) 
```

```python
X = iris.data.features 
y = iris.data.targets 
```

```python
X
```

```python
scaler = MinMaxScaler((-1, 1))
```

```python
scaler.fit(X)
```

```python
X_scaled = scaler.transform(X)
```

```python
X_scaled = pd.DataFrame(X_scaled, index=X.index, columns=X.columns)
```

```python
X_scaled
```

```python
y
```

```python
y_linsep = pd.Series([1 if x == "Iris-setosa" else 0 for x in y['class']])
```

```python
train_indices = sample(range(150), k=int(0.8 * 150))
```

```python
X_train = X_scaled.iloc[sorted(train_indices)]
```

```python
X_train
```

```python
X_test = X_scaled.iloc[~X_scaled.index.isin(X_train.index)]
```

```python
len(X_test)
```

```python
y_train = y_linsep[X_train.index]
```

```python
y_test = y_linsep[X_test.index]
```

```python
y_train
```

```python
from perceptron import Perceptron
p = Perceptron(4)
```

```python
losses = p.train(X_train, y_train, epochs=3)
```

```python
losses
```

```python
train_pred = p.test(X_train)
```

```python
test_pred = p.test(X_test)
```

```python
p.weights
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, test_pred)
```

```python
accuracy_score(y_train, train_pred)
```
