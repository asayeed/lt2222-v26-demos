```python
import spacy
```

```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
sentences = [review_df.iloc[0]['review'], review_df.iloc[1]['review']]
```

```python
sentences
```

```python
output = [nlp(x) for x in sentences]
```

```python
for token in enumerate(output[0]):
    print(token)
```

```python
def build_tagged_documents(documents, tagger):
    for i, document in enumerate(documents):
        processed = tagger(document)
        lemmas = [x.lemma_ for x in processed]
        yield TaggedDocument(lemmas, [i])
```

```python
tagged = build_tagged_documents(sentences, nlp)
```

```python
d2vmodel = Doc2Vec(list(build_tagged_documents(sentences, nlp)))
```

```python
len(d2vmodel.dv[0])
```

```python
import tqdm

all_reviews_tagged = []
for i, doc in tqdm.tqdm(enumerate(build_tagged_documents(list(review_df[0:1000]['review']), nlp))):
    all_reviews_tagged.append(doc)    
#all_reviews_tagged = list(build_tagged_documents(list(review_df['review']), nlp))
```

```python
reviews_d2vmodel = Doc2Vec(all_reviews_tagged)
```

```python
short_review_df = review_df[0:1000].copy()
```

```python
reviews_d2vmodel.dv[999]
```

```python
review_vectors = [reviews_d2vmodel[x] for x in range(1000)]
```

```python
short_review_df['vectors'] = review_vectors
```
