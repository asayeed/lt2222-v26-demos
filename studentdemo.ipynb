```python
student_mat = pd.read_csv("student-mat.csv", sep=";")
```

```python
student_mat
```

```python
avg_grade = student_mat['G3'].median()
```

```python
avg_grade
```

```python
student_mat['above_avg'] = (student_mat['G3'] >= avg_grade)
```

```python
student_mat
```

```python
student_mat['test_inst'] = np.random.randint(0, 100, len(student_mat))
```

```python
student_mat
```

```python
student_mat_train = student_mat[student_mat['test_inst'] < 80].copy()
```

```python
student_mat_test = student_mat[student_mat['test_inst'] >= 80].copy()
```

```python
from chefboost import Chefboost as chef
```

```python
config = {'algorithm': 'C4.5'}
```

```python
student_mat_train['above_avg'] = ['true' if x else 'false' for x in student_mat_train['above_avg']]
```

```python
student_mat_train
```

```python
student_mat_train.drop(['G1','G2','G3'], inplace=True, axis=1)
```

```python
student_mat_train
```

```python
config = {'algorithm': 'C4.5'}
model = chef.fit(student_mat_train, config = config, target_label = 'above_avg')
```

```python
student_mat_train.drop(['failures'], inplace=True, axis=1)
```

```python
config = {'algorithm': 'C4.5'}
model = chef.fit(student_mat_train, config = config, target_label = 'above_avg')
```

```python
student_mat_test['above_avg'] = ['true' if x else 'false' for x in student_mat_test['above_avg']]
```

```python
student_mat_test.drop(['G1','G2','G3','failures'], inplace=True, axis=1)
```

```python
student_mat_train.columns
```

```python
student_mat_test.columns
```

```python
chef.evaluate(model, student_mat_test, target_label='above_avg')
```
